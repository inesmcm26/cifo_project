```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

results = pd.read_csv('results.csv')
```

# Analysis of the best 5 best sets of hyperparameters

```python
# Apply the conversion function to all values in the DataFrame
results = results.applymap(eval)
```

## Best 5 algorithms

```python
new_results = pd.DataFrame()

for hyperparams in results.columns:
    new_results[hyperparams] = results[hyperparams].apply(np.median)
```

```python
new_results.head()
```

Let's see the median best value of every algorithm in the last generation:

```python
# Get the last row and sort its values in descending order
last_best_fitness = new_results.iloc[-1].sort_values(ascending=False)

# Print the column names and corresponding values in descending order
for column_name, value in last_best_fitness.items():
    print(f"{column_name}: {value}")
```

Retrieve the best 5 sets of hyperparameters

```python
best_hyperparams = new_results.tail(1).max().nlargest(5).index
```

```python
best_hyperparams
```

## Plot Evolutionary Trend

```python
results[best_hyperparams]
```

```python
def plot_evolutionary_trend(results, title, first_gen = 0, last_gen = 100):
    plt.figure(figsize=(8, 6))

    df = results.loc[first_gen:last_gen, :].copy()
    
    for hyperparams in df.columns:
        df[hyperparams + '_median'] = df[hyperparams].apply(np.median)
        df[hyperparams + '_std'] = df[hyperparams].apply(np.std)

    median_cols = df.filter(regex = 'median').columns
    stdev_cols = df.filter(regex = 'std').columns

    # Plot the averages with filled areas representing standard deviations
    for avg_col, std_col in zip(median_cols, stdev_cols):
        average = df[avg_col]
        std_dev = df[std_col]

        plt.plot(df.index, average, label=avg_col[:-7])
        plt.fill_between(df.index, average - std_dev, average + std_dev, alpha=0.2)

    plt.xlabel('Generation')
    plt.ylabel('Fitness')

    plt.title(title)
    plt.legend(fontsize = 9, loc = 'lower right')
    plt.show()

plot_evolutionary_trend(results.loc[:, best_hyperparams], title = 'Evolutionary Trend of the Best Algorithms')
```

```python
plot_evolutionary_trend(results.loc[:, best_hyperparams], title = 'Evolutionary Trend of the Best Algorithms (after 10 Generation)',
                        first_gen = 10, last_gen = 100)
```

# Analysis of the impact of crossover

```python
crossover_methods = ['eager_breeder', 'gbx', 'twin_maker']

df_crossover = pd.DataFrame()

results_copy = results.copy()

for crossover in crossover_methods:

    # Calculate median performance in each generation for each crossover method
    for hyperparams in results.filter(regex = crossover).columns:

        results_copy.loc[:, hyperparams] = results_copy.loc[:, hyperparams].apply(np.median)

    # Save the performances of each algorithm that uses each crossover in a list
    df_crossover[crossover] = results_copy.filter(regex = crossover).apply(lambda row: row.tolist(), axis=1)

df_crossover
```

```python
plot_evolutionary_trend(df_crossover, title = 'Evolutionary Trend Among Crossover Methods')
```
