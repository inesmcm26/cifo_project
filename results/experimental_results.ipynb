```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the Grid Search results
results = pd.read_csv('results.csv')
```

# 1. Analysis of the best 5 best sets of hyperparameters

```python
results = results.applymap(eval)
```

The _results_ dataframe consists of one column for each set of tested hyperparameters and one row for each generation. The cells within the dataframe contain lists that represent the highest fitness values discovered by each algorithm in each run of each generation.

```python
results.head()
```

Now, we apply the median to each list, to get the median best fitness in each generation for each algorithm.

```python
new_results = pd.DataFrame()

for hyperparams in results.columns:
    new_results[hyperparams] = results[hyperparams].apply(np.median)
```

```python
new_results.head()
```

Let's see the median best value of every algorithm in the last generation.

```python
# Get the last row and sort its values in descending order
last_best_fitness = new_results.iloc[-1].sort_values(ascending=False)

# Print the column names and corresponding values in descending order
for column_name, value in last_best_fitness.items():
    print(f"{column_name}: {value}")
```

Now we can retrieve the best 5 sets of hyperparameters.

```python
best_hyperparams = new_results.tail(1).max().nlargest(5).index
```

```python
best_hyperparams
```

After finding the 5 algorithms with highest fitness in the last generation, we can analyse each evolutionary trend.

```python
# Slice of the dataframe with the best hyperparameters
results[best_hyperparams]
```

```python
def plot_evolutionary_trend(results, title, first_gen = 0, last_gen = 100):
    """
    Function that plots the evolutionary trend of GA algorithms

    Args:
        results (pd.DataFrame): dataframe with a list of fitnesses for each generation and hyperparameters
        title (str): title of the plot
        first_gen (int): first generation to plot
        last_gen (int): last generation to plot
    """
    plt.figure(figsize=(8, 6))

    # Slice the dataframe in the desired generations
    df = results.loc[first_gen:last_gen, :].copy()
    
    # Calculate the median and standard deviation of each generation
    for hyperparams in df.columns:
        df[hyperparams + '_median'] = df[hyperparams].apply(np.median)
        df[hyperparams + '_std'] = df[hyperparams].apply(np.std)

    # Get the columns with the median and standard deviation
    median_cols = df.filter(regex = 'median').columns
    stdev_cols = df.filter(regex = 'std').columns

    # Plot the medians with filled areas representing standard deviations
    for avg_col, std_col in zip(median_cols, stdev_cols):
        average = df[avg_col]
        std_dev = df[std_col]

        plt.plot(df.index, average, label=avg_col[:-7])
        plt.fill_between(df.index, average - std_dev, average + std_dev, alpha=0.2)

    plt.xlabel('Generation')
    plt.ylabel('Fitness')

    plt.title(title)
    plt.legend(fontsize = 9, loc = 'lower right')
    plt.show()

plot_evolutionary_trend(results.loc[:, best_hyperparams], title = 'Evolutionary Trend of the Best Algorithms')
```

And we can further analyse the lines after performance stabilizes

```python
plot_evolutionary_trend(results.loc[:, best_hyperparams], title = 'Evolutionary Trend of the Best Algorithms (after 10 Generation)',
                        first_gen = 10, last_gen = 100)
```

# 2. Analysis of the impact of crossover

To assess the impact of each crossover method, we plot the median performance of all algorithms that employed each type of crossover, along with their corresponding standard deviations.

```python
crossover_methods = ['eager_breeder', 'gbx', 'twin_maker']

df_crossover = pd.DataFrame()

results_copy = results.copy()

for crossover in crossover_methods:

    # Get all algorithms that use the same crossover method
    for hyperparams in results.filter(regex = crossover).columns:

        # Calculate median performance in each generation for each crossover method
        results_copy.loc[:, hyperparams] = results_copy.loc[:, hyperparams].apply(np.median)

    # Save the performances of each algorithm that uses each crossover in a list
    df_crossover[crossover] = results_copy.filter(regex = crossover).apply(lambda row: row.tolist(), axis=1)

# Dataframe with one column for each crossover method
# Each cell is a list of the median fitnesses of each generation
# for each algorithm that uses that crossover method
df_crossover
```

```python
plot_evolutionary_trend(df_crossover, title = 'Evolutionary Trend Among Crossover Methods')
```
